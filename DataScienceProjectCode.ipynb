```python
import pandas as pd
```

```python
#initializing the two data sets
left = pd.read_csv("Ratings.csv")
right = pd.read_csv("Sales.csv")
```

```python
#shorten platform on left list to prepare for the merge
left['platform'].replace({"Dreamcast":"DC",
                          "Game Boy Advance":"GBA",
                          "GameCube":"GC",
                          "Nintendo 64":"N64",
                          "PlayStation":"PS",
                          "PlayStation 2":"PS2",#PS replacements left a weird space inbetween ps and the number, like 'PS 3'
                          "PlayStation 3":"PS3",
                          "PlayStation 4":"PS4",
                          "PlayStation Vita":"PSV",
                          "Xbox 360":"X360",
                          "Xbox One":"XOne",},regex=True, inplace=True)
#p.s. This replace thing took me way longer to get working than I would ever care to admit...
left['platform'].replace({"Xbox":"XB"," ":""},regex=True, inplace=True) #quick fix to PS issue by removing space
left
```

```python
#Renamed game name and platform column to match and merge two tables together
left.rename(columns={"name": "Name"}, inplace=True)
left.rename(columns={"platform": "Platform"}, inplace=True)
mergedtable = pd.merge(right, left, on=["Name","Platform"])
mergedtable
#Did this to combine the data of both the game's ratings and the game's sales.
#Table merged but unclean
```

```python
#Cleaning the table up by removing irrelevant information to make it somewhat more readable
newtable = mergedtable.drop(columns = ['summary', 'release_date', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'])
newtable
```

```python
newtable['Year_of_Release'] = newtable['Year_of_Release'].fillna(2006.0)
#All games with unknown release dates will default to 2006, cuz when I think of a bad year for video games, Sonic '06 comes to mind...
ReleaseYear = newtable.sort_values(by=['Year_of_Release'], ascending=False)
ReleaseYear
```

```python
#Create new rudimentary category to measure game success
def MeasureSuccess(row):
    if (row["Rating"]=="E" or row["Rating"]=="E10+"):
        if row["Global_Sales"]>=15:
            return "YES"
        elif (row["meta_score"]>=75):
            return "YES"
        else:
            return "NO"
    elif row["Rating"]=="M":
        if row["Global_Sales"]>=10:
            return "YES"
        elif row["meta_score"]>=80:
            return "YES"
        else:
            return "NO"
    else:
        if row["Global_Sales"]>=15:
            return "YES"
        elif row["meta_score"]>=75:
            return "YES"
        else:
            return "NO"
        
newtable["Success?"] = newtable.apply(lambda row: MeasureSuccess(row), axis=1)
newtable.head(500)
```

```python
#machine learning portion
import sys
import matplotlib
matplotlib.use('Agg')
import numpy
from sklearn import linear_model


from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

RatingsKey = {"AO":0,
              "E":1,
              "E10+":1,
              "EC":1,
              "K-A":1,
              "M":0,
              "RP":1,
              "T":1}
newtable['Rating']=newtable['Rating'].map(RatingsKey)
newtable['Rating']=newtable['Rating'].fillna(1)
SuccessKey = {'YES': 1, 'NO': 0}
newtable['Success?']=newtable['Success?'].map(SuccessKey)
newtable['Success?']=newtable['Success?'].fillna(0)

newtable.head(20)

features = ['Rating','meta_score','Global_Sales']

X = newtable[features]
Y = newtable['Success?']

#Decision Tree
dtree = DecisionTreeClassifier()
X = X.values
dtree = dtree.fit(X, Y)

tree.plot_tree(dtree, feature_names=features)

#prediction with a game rated M, had a meta score of 81, and sold 9 million copies
print(dtree.predict([[0, 81, 9.1]]))
#[0] = no, [1] = yes
```

```python
#Linear Regression
X = newtable[features]
Y = newtable['Success?']

logr = linear_model.LogisticRegression()
X = X.values
logr.fit(X,Y)
#prediction with a game rated M, had a meta score of 78, and sold 9 million copies
predicted = logr.predict(numpy.array([[0, 78, 9.1]]))
print(predicted)
```
